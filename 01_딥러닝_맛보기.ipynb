```python
# 딥러닝을 구동하는데 필요한 라이브러리
from keras.models import Sequential
from keras.layers import Dense

import pandas as pd
import tensorflow as tf
```

## 수술 환자 데이터 가져오기
- 1이 생존, 0이 사망

```python
df = pd.read_csv('dataset/ThoraricSurgery.csv', header=None)
df
```

## 입력(x)과 결과(y)로 나누기

```python
x = df.drop(17, axis=1)
y = df[17]
```

## 딥러닝 구조를 결정한다. (모델을 설정하고 실행하는 부분)

```python
# 데이터의 랜덤성을 고정한다.
tf.random.set_seed(1)

model = Sequential()
# 2개의 layer를 만듬 
# input_dim 입력되는 컬럼갯수
model.add(Dense (30, input_dim=17, activation='relu')) # 입력 레이어
model.add(Dense (1, activation='sigmoid')) # 출력 레이어
```

## 딥러닝 실행, 학습부분

```python
# loss : 오차를 계산하는 방식 (mean_squared_error : 선형모델에서 진짜 결과와 예측 결과를 계산하는 함수)
# optimizer : 값을 보정하는 함수
# metrics : 평가방법
model.compile(loss="mean_squared_error", optimizer='adam',  metrics =['accuracy'])

# 학습
# epochs : 학습과 보정 반복횟수
model.fit(x,y,epochs=30, batch_size =10)
```

## 예측


```python
# - predict : (예측 결과가 1일 확률) 

pred = model.predict(x)
```

```python
# 값으로 나오게끔 하기
result= (pred>0.5).astype(int)
```

```python
from sklearn.metrics import accuracy_score

score = accuracy_score(y, result)
score
```
