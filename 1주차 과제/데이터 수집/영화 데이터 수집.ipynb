```python
# 분석 및 수집
from bs4 import BeautifulSoup
# 요청
import requests

import urllib.request

import os
import pandas as pd
import numpy as np
```

## 요청하는 함수

```python
def getConnection(site):
    # 요청
    response = requests.get(site)
    
    # 응답 결과를 가지고 bs4 객체를 생성한다.
    soup = BeautifulSoup(response.text, 'lxml')
    return soup
```

```python
a='https://movie.daum.net/premovie/released?reservationOnly=N&sort=reservation&page=1'
soup = getConnection(a)
soup
```

## 분석 및 수집, 저장하는 함수

```python
title1 = []
score1 = []
date1  = []
rate1  = []

# 한 페이지에서 추출하고자 하는 데이터를 모두 추출해 저장한다.
def getData(soup,j):
    # 제목
    titles = soup.find_all('a',class_='link_g')
    for title in titles:
       title1.append(title.get_text())
    
    # 평점
    scores = soup.find_all('em',class_="emph_grade")
    for score in scores:
        score1.append(score.get_text())
        print(score)
        
    # 개봉일
    dates = soup.find_all('dd')
    for i in range(0,len(dates),2):
        date1.append(dates[i].get_text())
        
    # 예매율
    rates = soup.find_all('dd')
    for i in range(1,len(rates),2):
        rate1.append(rates[i].get_text())
        
    # 이미지 수집
    images = soup.find_all('img',class_="lazy thumb_photo")

    for i in range(0,len(images)):
        images = soup.find_all('img',class_="lazy thumb_photo")[i]['src']
        images = 'http:' + images
        print(images)
        urllib.request.urlretrieve(images,f'{i}.jpg')
        if os.path.exists(f'{i}.jpg'):
            urllib.request.urlretrieve(images,f'{i+(j+1)*10}.jpg')
    
    
```

```python
getData(soup,1)
```

```python
# 예매율
# 페이지 갯수 구하기
pages = soup.find('div',class_='inner_paging')
b= pages.get_text()
total_pages= b.split("\n")[1:-1]
len(total_pages)
```

```python
# 다음 페이지 존재 여부 확인
def getNextPage(soup):
    for i in range(0,len(total_pages)+1):
        a=f'https://movie.daum.net/premovie/released?reservationOnly=N&sort=reservation&page={i}'
        soup = getConnection(a)
        getData(soup,i)
```
