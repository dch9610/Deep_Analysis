```python
# seed 설정
np.random.seed(3)
tf.random.set_seed(3)
```

```python
# 데이터 불러오기
df1 = pd.read_csv('../dataset/wine.csv', header=None)
df1.head()
```

```python
# 데이터 분류
x = df1.drop(12, axis=1)
y = df1[12]
```

```python
# 모델 설정
model = Sequential()
model.add(Dense(30, input_dim=12, activation='relu'))
model.add(Dense(12, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
# 모델 컴파일
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# 자동 중단설정
# patience : 지정해준 숫자만큼 동안 성능향상이 이루어지지 않았다면 중단
early_stopping_callback = EarlyStopping(monitor='val_loss', patience =1000 )
```

```python
# 저장될 위치와 파일명
model_path = 'models/{epoch} - {val_loss}.hdf5'

# 저장 조건
# model_path에 저장함
# monitor = 'val_loss' 오차, 'val_accuracy' 정확도
# save_best_only = 최적의 값이 나오면 더 이상 저장하지않음
checkpointer = ModelCheckpoint(filepath = model_path, monitor='val_loss', save_best_only = True)
```

```python
# 학습
# 하나의 epoch이 끝나면 callbacks가 실행
model.fit(x, y, validation_split=0.2, epochs=2000, batch_size=200, verbose=1, callbacks=[early_stopping_callback,checkpointer])
```

```python
# 학습완료된 모델을 로딩한다.
best_model = load_model('models/96 - 0.027031388133764267.hdf5')
best_model
```

```python
# 예측결과 추출한다.
pred = (best_model.predict(x)>0.5).astype('int32')
pred
```

```python
r1 = accuracy_score(y, pred)
r1
```
