```python
# seed 설정
np.random.seed(3)
tf.random.set_seed(3)
```

```python
# 데이터 불러오기
df1 = pd.read_csv('data/wine.csv', header=None)
df1.head()
```

```python
# 데이터 프레임의 일부를 발췌한다.
# df1 = df1.sample(n=1000).reset_index(drop=True)
# df1
```

```python
# 데이터 분류
x = df1.drop(12, axis=1)
y = df1[12]

display(x)
display(y)
```

```python
# 모델 설정
model = Sequential()
model.add(Dense(30, input_dim=12, activation='relu'))
model.add(Dense(12, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
# 모델 컴파일
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
#  저장될 위치와 파일명
modelpath = 'models/{epoch}-{val_accuracy}.hdf5'
# 저장 조건
checkpointer = ModelCheckpoint(filepath=modelpath, monitor='val_accuracy', save_best_only=True)
```

```python
# 학습
model.fit(x, y, validation_split=0.2, epochs=200, batch_size=200, verbose=1, callbacks=[checkpointer])
```

```python
# 모델을 가져온다.
best_model = load_model('models/104-0.9961538314819336.hdf5')
best_model
```

```python
best_model.evaluate(x, y)[1]
```
