```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 데이터
x_data = [2,4,6,8,10,12,14]
y_data = [0,0,0,1,1,1,1]
```

```python
# 그래프로 나타내기
plt.scatter(x_data, y_data)
plt.show()
```

```python
# a와 b의 값
a = 0
b = 0
```

```python
# 학습률
lr = 0.05
```

```python
# 시그모이드 함수
def sigmoid(x):
    return 1 / (1+ np.e ** (-x))
```

```python
# 반복 횟수
epochs = 100001
```

```python
# ndarray 생성
x_data1 = np.array(x_data)
y_data1 = np.array(y_data)
```

```python
# 경사 하강법
for i in range(epochs):
    # 예측 결과를 가져온다.
    y_pred = sigmoid(a * x_data1 + b)
    
    # 오차를 계산
    error = y_pred - y_data1
    
    # a와 b에 대해 편미분을 수행한다.
    a_diff = (1/len(x_data1)) * sum(x_data1 * error)
    b_diff = (1/len(x_data1)) * sum(error)
    
    # a와 b의 값을 계산한다.
    a = a - lr * a_diff
    b = b - lr * b_diff
    
    if i % 10000 == 0:
        print(f'epochs : {i}, a : {a}, b : {b}')
        
        plt.scatter(x_data1, y_data1)

        # 그래프로 나타낼 x값의 범위
        x_range = (np.arange(0, 15,0.1))
        plt.plot(x_range, np.array([sigmoid(a*x+b) for x in x_range]))
        plt.show()
    
```

```python
# 예측결과 (0보다 작은 값은 0, 1보다 큰 값은 1)
y_pred = a * x_data1 + b
y_pred
```
