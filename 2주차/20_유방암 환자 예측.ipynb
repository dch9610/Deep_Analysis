```python
# seed 설정
np.random.seed(3)
tf.random.set_seed(3)
```

```python
# 데이터 불러오기
df1 = pd.read_csv('data/breast_cancer.csv')
df1.head()
```

```python
# 데이터 분류
x = df1.drop('target', axis=1)
y = df1['target']

display(x)
display(y)
```

```python
# 문자열을 숫자로 변환
encoder1 = LabelEncoder()
encoder1.fit(y)
y = encoder1.transform(y)
y
```

```python
# 모델 설정
model = Sequential()
model.add(Dense(60, input_dim=30, activation='relu'))
model.add(Dense(40, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
# 모델 컴파일
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# 저장 콜백
# modelpath = 'models/accuracy_{epoch}-{val_accuracy}.hdf5'
# callback1 = ModelCheckpoint(filepath=modelpath, monitor='val_accuracy', verbose=1, save_best_only=True)

modelpath = 'models/loss_{epoch}-{val_loss}.hdf5'
callback1 = ModelCheckpoint(filepath=modelpath, monitor='val_loss', verbose=1, save_best_only=True)
```

```python
# 조기 중단 콜백
# callback2 = EarlyStopping(monitor='val_accuracy', patience=1000)
callback2 = EarlyStopping(monitor='val_loss', patience=1000)
```

```python
# 학습
model.fit(x, y, validation_split=0.2, epochs=100000, batch_size=50, 
         callbacks=[callback1, callback2])
```

```python
# 모델 불러오기
loss_model = load_model('models/loss_2479-0.00948227196931839.hdf5')
acc_model = load_model('models/accuracy_499-0.9890109896659851.hdf5')
```

```python
# 학습한 데이터로 성능평가
a1 = loss_model.evaluate(x, y)[1]
a2 = acc_model.evaluate(x, y)[1]

print(a1)
print(a2)
```

```python
# 예측할 데이터를 불러온다.
df2 = pd.read_csv('data/breast_cancer_new.csv')
df2.head()
```

```python
# 예측 결과를 추출한다.
pred = (loss_model.predict(df2) > 0.5).astype('int32')
pred
```

```python
# 결과를 문자열로 복원한다.
pred2 = encoder1.inverse_transform(pred)
pred2
```

```python
# 결과 데이터를 붙힌다.
df2['target'] = pred2
df2.head()
```

```python
df2.to_csv('data/breast_cancer_deep_result.csv')
```
